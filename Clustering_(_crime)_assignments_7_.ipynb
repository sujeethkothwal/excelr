```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import KMeans
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
```

```python
c = pd.read_csv('crime_data.csv')
c.head()
```

```python
crime=c.rename({'Unnamed: 0':'States'},axis=1)
crime.tail()
```

```python
crime.States.value_counts()
```

```python
crime.info()
```

```python
crime.isna().sum()
```

```python
#hierarchy Clustering
n = MinMaxScaler()
data= n.fit_transform(crime.iloc[:,1:].to_numpy())
crimes=pd.DataFrame(data, columns = crime.columns[1:])
crimes.head()
```

```python
#Normalization function
def norm_func(i):
  x=(i-i.min())/(i.max()-i.min())
  return(x)


#Normalized data frame (considering the numerical part of data)
crime_norm = norm_func(crime.iloc[:,1:])

crime_norm
```

#Clustering

```python
#create dendogram
fig =plt.figure(figsize=(15,8))
dendrogram = sch.dendrogram(sch.linkage(crimes,method= 'average'))
```

```python
#create cluster
hc1=AgglomerativeClustering(n_clusters=4,affinity = 'euclidean',linkage='average')
```

```python
y2=hc1.fit_predict(crimes)
cc=pd.DataFrame(y2,columns=['Clusters'])
cc.head()
```

```python
crimes2 = pd.concat([crime,cc],axis=1)
crimes2.head(7)
```

```python
crimes2.sort_values("Clusters").reset_index()
```

```python
crimes2['Clusters'].value_counts()
```

#kmeans

```python
kmeans=KMeans(n_clusters=4,random_state=0)
kmeans.fit(crimes)
```

```python
kmeans.inertia_
```

```python
wcss=[]
for i in range(1, 8):
  kmeans= KMeans(n_clusters=i,random_state=0)
  kmeans.fit(crimes)
  wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1,8),wcss)
plt.title('Eblow graph')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
#Build Cluster algorithm
cc2=KMeans(4, random_state=8)
cc2.fit(crimes)
```

```python
cc2.labels_
```

```python
df2=pd.DataFrame(cc2.labels_,columns=['clusters'])
```

```python
crimes3= pd.concat([crimes,df2],axis=1)
crimes3.head()
```

```python
crimes3['clusters'].value_counts()
```

```python
crimes3.groupby('clusters').agg(['mean'])
```

#DBSCAN

```python
array=crimes.values
```

```python
stscaler=StandardScaler().fit(array)
X2=stscaler.transform(array)
```

```python
X2
```

```python
dbscan=DBSCAN(eps=0.98,min_samples=3)
dbscan.fit(X2)
```

```python
dbscan.labels_
```

```python
c2=pd.DataFrame(dbscan.labels_,columns= ['Cluster ID'])
c2.value_counts()
```

```python
crimes4 = pd.concat([crime,c2],axis=1)
crimes4
```
